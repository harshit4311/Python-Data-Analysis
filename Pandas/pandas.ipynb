```python
import pandas as pd

print('import done')
```

```python
# series 
''' 1D array-like object which can hold any data-type.
--> similar to a column in a table '''

data = [1, 2, 3, 4, 5]

series = pd.Series(data)
print(series)
print(type(series))
```

```python
# series from a dictionary 
data = {'a':1, 'b':2, 'c':3}

series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10, 20, 30]
index = ['a', 'b', 'c']

pd.Series(data, index=index)
```

```python
# dataframe
''' dataframe from a dictionary of list '''

data = {
    'Name': ['Harshit', 'Krish', 'John'],
    'Age': [19, 35, 50],
    'City': ['London', 'NYC', 'Palo alto']    
}

df = pd.DataFrame(data)
df
```

```python
print(type(df))
```

```python
''' dataframe from a list of dictionaries '''
data = [
    {'Name': 'Harshit', 'Age': 19, 'City': 'London'},
    {'Name': 'Krish', 'Age': 35, 'City': 'NYC'},
    {'Name': 'John', 'Age': 50, 'City': 'Palo Alto'}
]

df = pd.DataFrame(data)
df
```

```python
print(type[df])
```
